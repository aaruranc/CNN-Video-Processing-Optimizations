```python
import pandas as pd
file = 'BsmP1_10204_1498.csv'
headers = ['RxDevice', 'FileID', 'TxDevice', 'Gentime', 'TxRandom', 'MsgCount', 'DSecond', 
           'Latitude', 'Longitude', 'Elevation', 'Speed', 'Heading', 'Ax', 'Ay', 'Az', 
           'Yawrate', 'PathCount', 'RadiusOfCurve', 'Confidence']
df = pd.read_csv(file, names=headers)
df
```

```python
timeDiff = []
for k in range(1, len(df.index)):
    timeDiff.append((df['Gentime'][k] - df['Gentime'][k-1])/1000000)
    
timeDiff
```

```python
totalLength = (df['Gentime'][len(df.index)-1] - df['Gentime'][0])/1000000
latitudeChange = df['Latitude'][len(df.index)-1] - df['Latitude'][0]
longitudeChange = df['Longitude'][len(df.index)-1] - df['Longitude'][0]
elevationChange = df['Elevation'][len(df.index)-1] - df['Elevation'][0]

print('total time taken is ' + str(totalLength) + ' seconds')
print('total latitude change is ' + str(latitudeChange) + ' degrees')
print('total latitude change is ' + str(longitudeChange) + ' degrees')
print('total elevation change is ' + str(elevationChange) + ' meters')
```
