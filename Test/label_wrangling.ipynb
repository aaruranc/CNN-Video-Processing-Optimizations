```python
import pandas as pd
file = 'BsmP1_10204_01834.csv'
headers = ['RxDevice', 'FileID', 'TxDevice', 'Gentime', 'TxRandom', 'MsgCount', 'DSecond', 
           'Latitude', 'Longitude', 'Elevation', 'Speed', 'Heading', 'Ax', 'Ay', 'Az', 
           'Yawrate', 'PathCount', 'RadiusOfCurve', 'Confidence']
df = pd.read_csv(file, names=headers)
df
```

```python
file = 'classified_images_trimmed.csv'
df2 = pd.read_csv(file)
df2
```

```python
# 2524 - 15290 Frames are selected to match up to BSM csv
```

```python
count = 0
cols = list(df)
df3 = pd.DataFrame(columns = cols)

for index, row in df2.iterrows():
    name = df2['Image'][index]
    detected = df2['NumDetected'][index]
    
    if pd.isna(name):
        continue
    else:
        detected = int(detected)
        if detected > 1:
            df3 = df3.append(df.iloc[count])
            detected = detected - 1
            while detected !=0:
                df3 = df3.append(pd.Series(), ignore_index=True)
                detected = detected - 1
        elif detected == 0 or detected == 1:
            df3 = df3.append(df.iloc[count])
        count = count + 1
        
df3
        
```

```python
df4 = pd.concat([df3, df2], axis=1, join_axes=[df3.index])
headers = ['RxDevice', 'FileID', 'TxDevice', 'Gentime', 'TxRandom', 'MsgCount', 'DSecond', 
           'Latitude', 'Longitude', 'Elevation', 'Speed', 'Heading', 'Ax', 'Ay', 'Az', 
           'Yawrate', 'PathCount', 'RadiusOfCurve', 'Confidence', 'Image', 'NumDetected', 
           'Class', 'Score', 'xmin', 'xmax', 'ymin', 'ymax']
df4 = df4[headers]
df4
```

```python
df4.to_csv('structured_data.csv', index=False)
```
